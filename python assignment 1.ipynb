1. Build a Personal Finance Tracker using Python that: 
a. Accepts income & multiple expenses 
b. Calculates savings & percentage spent 
c. Categorizes spending: LOW, MODERATE, HIGH 
d. Suggests message: "Good control!", "Be careful!", or "Financial Risk!" 
e. Perform exception handling in case of any exception geneated during program 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

income = float(input("Enter your monthly income: "))
n = int(input("How many expenses do you have? "))

amounts = []
names = []

for i in range(n):
    name = input(f"Expense {i+1} name: ")
    amt = float(input("Amount: "))
    names.append(name)
    amounts.append(amt)

df = pd.DataFrame({"Expense": names, "Amount": amounts})

total_exp = df["Amount"].sum()
savings = income - total_exp
percent = (total_exp / income * 100) if income > 0 else 0

if percent < 40:
    level = "LOW"
    msg = "Good Control!"
elif percent < 75:
    level = "MODERATE"
    msg = "Be careful!"
else:
    level = "HIGH"
    msg = "Financial Risk!"

print("\n--- REPORT ---")
print(df)
print("Total Expense:", total_exp)
print("Savings:", savings)
print("Percent Spent:", round(percent, 2), "%")
print("Category:", level)
print("Suggestion:", msg)

df.plot(kind="bar", x="Expense", y="Amount")
plt.title("Your Expenses")
plt.xlabel("Expense")
plt.ylabel("Amount")
plt.show()
```

2.  Describe the purpose of metacharacters like ., ^, $, [], (), and |. Provide a valid use 
case for each. 
a. What are quantifiers in Regex? Explain *, +, ?, and {m,n} with suitable 
matching examples. Using regular expression extract the informations such as:  
b. Indian Mobile No. 
c. PAN Card 
d. Aadhaar 
e. Postal PIN 
f. URL 
g. IPv4 
h. Password (Strong) 

```python
import re

text = """
Phone : 9876543210
PAN : ABCDE1234F
Aadhar : 1234 5678 9012
PIN : 560001
URL : https://example.com/page
IP : 192.168.1.10
Password : Abc@1234
"""

patterns = {
    "Indian Mobile": r"(?:\+91[\s-]?|0)?[6-9]\d{9}",
    "PAN": r"[A-Z]{5}\d{4}[A-Z]",
    "Aadhaar": r"\b\d{4}\s\d{4}\s\d{4}\b",
    "Postal PIN": r"[1-9][0-9]{5}",
    "URL": r"https?://[^\s]+",
    "IPv4": r"\b(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)"
            r"(?:\.(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)){3}\b",
    "Strong Password": r"(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@$!%*?&]).{8,}"
}

for name, pat in patterns.items():
    print(name, ":", re.findall(pat, text))
```

3. Write a Python program to create and maintain a COVID vaccination record file. 
 Features: 
a. Append new records into a text file 
b. Read and display all records aligned in columns 
c. Ensure that no record is overwritten, only appended 
d. Explain why persistent file storage is essential in real-world systems 

```python
import pandas as pd
import os

file = "vaccination_records.csv"

if os.path.exists(file) and os.path.getsize(file) > 0:
    df = pd.read_csv(file)
else:
    df = pd.DataFrame(columns=["Name", "Age", "Gender", "Vaccine", "Dose", "Date"])

while True:
    print("\n1: Add Record\n2: Show Records\n3: Exit")
    choice = input("Enter choice: ").strip()

    if choice == "1":
        name = input("Name: ").strip()
        age = input("Age: ").strip()
        gender = input("Gender: ").strip()
        vaccine = input("Vaccine Name: ").strip()
        dose = input("Dose Number: ").strip()
        date = input("Date (YYYY-MM-DD): ").strip()

        df.loc[len(df)] = [name, age, gender, vaccine, dose, date]
        df.to_csv(file, index=False)
        print("Record Added Successfully.")

    elif choice == "2":
        print("\n--- VACCINATION RECORDS ---")
        if df.empty:
            print("No records found.")
        else:
            print(df.to_string(index=False))

    elif choice == "3":
        print("Exiting...")
        break

    else:
        print("Invalid choice! Please enter 1, 2, or 3.")
```

4. Create a Library Management System using lists, dict, sets, file handling, and 
regex. Functionalities: 
a. Add new books with ID, name, author, category 
b. Search books using regex keyword 
c. Prevent duplicate book IDs using set() 
d. Update or delete book info stored inside a dictionary 
e. Save and read all book records from a file 
f. Write 3–4 lines explaining how multiple data structures work together 
effectively.

```python
import pandas as pd
import re
import os

file = "library_books.csv"

if os.path.exists(file) and os.path.getsize(file) > 0:
    df = pd.read_csv(file)
else:
    df = pd.DataFrame(columns=["ID", "Name", "Author", "Category"])

id_set = set(df["ID"].astype(str))

while True:
    print("\n1: Add Book  2: Search  3: Update  4: Delete  5: Show All  6: Exit")
    ch = input("Enter choice: ").strip()

    if ch == "1":
        bid = input("Book ID: ").strip()

        if bid in id_set:
            print("Duplicate ID not allowed!")
            continue

        name = input("Book Name: ").strip()
        author = input("Author: ").strip()
        cat = input("Category: ").strip()

        df.loc[len(df)] = [bid, name, author, cat]
        id_set.add(bid)
        df.to_csv(file, index=False)
        print("Book Added.")

    elif ch == "2":
        key = input("Enter keyword or regex: ").strip()

        result = df[df.apply(lambda row:
                             bool(re.search(key, " ".join(map(str, row.values)), re.I)),
                             axis=1)]

        if result.empty:
            print("No matching books found.")
        else:
            print("\n--- Search Results ---")
            print(result.to_string(index=False))

    elif ch == "3":
        bid = input("Enter ID to update: ").strip()
        row = df[df["ID"] == bid]

        if row.empty:
            print("Book not found!")
        else:
            idx = row.index[0]

            new_name = input("New Name (leave blank to keep): ").strip()
            new_author = input("New Author (leave blank to keep): ").strip()
            new_cat = input("New Category (leave blank to keep): ").strip()

            if new_name:
                df.at[idx, "Name"] = new_name
            if new_author:
                df.at[idx, "Author"] = new_author
            if new_cat:
                df.at[idx, "Category"] = new_cat

            df.to_csv(file, index=False)
            print("Book updated.")

    elif ch == "4":
        bid = input("ID to delete: ").strip()

        if bid not in id_set:
            print("Book ID not found!")
            continue

        df = df[df["ID"] != bid]
        id_set.discard(bid)

        df.to_csv(file, index=False)
        print("Book deleted.")

    elif ch == "5":
        if df.empty:
            print("No books in the library.")
        else:
            print("\n--- ALL BOOKS ---")
            print(df.to_string(index=False))

    elif ch == "6":
        print("Exiting...")
        break

    else:
        print("Invalid choice! Enter a number between 1–6.")
```
