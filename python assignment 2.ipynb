1. Python is considered an object-oriented programming language, but it is often said that 
Python supports OOP concepts such as inheritance, encapsulation, abstraction in a loo 
se and flexible manner compared to strictly OOP-based languages such as Java or 
C++. Explain this statement with justification. Write a Python program demonstrating

```python
class Animal:
    def __init__(self, name):
        self.name = name    

    def speak(self):
        raise NotImplementedError("Subclasses should implement this!")

class Dog(Animal):
    def speak(self):
        return f"{self.name} says: Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says: Meow!"
        
def animal_sound(animal):
    print(animal.speak())

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  

    # Getter
    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Deposit amount must be positive")

if __name__ == "__main__":
    d = Dog("Rex")
    c = Cat("Mittens")
    animal_sound(d) 
    animal_sound(c)  

    acct = BankAccount("Tushar", 1000)
    print("Balance:", acct.get_balance())
    acct.deposit(500)
    print("Balance after deposit:", acct.get_balance())

    d.favorite_toy = "Ball"
    print(d.name, "likes", d.favorite_toy)
```

2. Ride-sharing apps (Uber/Ola) calculate trip fares differently for Auto, Bike, Mini, 
Sedan, SUV categories. Explain how polymorphism is preferred instead of multiple 
if-else statements by writing Python code showing polymorphic calculate_fare() for 
at least three vehicle types. (Code Only) 

```python

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def calculate_fare(self, distance_km):
        pass

class Auto(Vehicle):
    def calculate_fare(self, distance_km):
        base = 25         
        per_km = 8       
        return base + per_km * distance_km

class Bike(Vehicle):
    def calculate_fare(self, distance_km):
        base = 10
        per_km = 5
        surge = 1.0      
        return base + per_km * distance_km * surge

class Sedan(Vehicle):
    def calculate_fare(self, distance_km):
        base = 40
        per_km = 12
        luggage_fee = 0
        if distance_km > 10:   
            luggage_fee = 20
        return base + per_km * distance_km + luggage_fee

if __name__ == "__main__":
    vehicles = [Auto(), Bike(), Sedan()]
    distance = 7.5  # km
    for v in vehicles:
        print(v.__class__.__name__, "fare for", distance, "km =", v.calculate_fare(distance))
```

3. Hospitals manage different types of medical staff like Doctor, Nurse, Surgeon, and 
Lab Technician who share some common details but also have role-specific 
functions.Inheritance avoids code duplication and improves maintainability in 
such systems. Write a Python program using a base class Staff and derived classes 
with unique behaviors. (Code only)

```python
class Staff:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def details(self):
        return f"Name: {self.name}, ID: {self.id}"

class Doctor(Staff):
    def duty(self):
        return f"Dr. {self.name} treats patients."

class Nurse(Staff):
    def duty(self):
        return f"Nurse {self.name} assists doctors."

class Technician(Staff):
    def duty(self):
        return f"Technician {self.name} performs lab tests."


d = Doctor("Anita", 101)
n = Nurse("Ravi", 102)
t = Technician("Priya", 103)

print(d.details(), "-", d.duty())
print(n.details(), "-", n.duty())
print(t.details(), "-", t.duty())
```

4.  Banking applications do not allow direct access to Account Balance, PIN, or Credit 
Card Limit. Explain how encapsulation protects sensitive data and allows 
controlled access using getter/setter methods. Write Python code using private or 
protected variables to demonstrate this. (Code Only)

```python
class Account:
    def __init__(self, acc_no, pin, balance):
        self.acc_no = acc_no
        self.__pin = pin
        self.__balance = balance

    def check_balance(self, pin):
        if pin == self.__pin:
            return self.__balance
        return "Incorrect PIN"

    def withdraw(self, amount, pin):
        if pin == self.__pin and amount <= self.__balance:
            self.__balance -= amount
            return "Withdrawal successful"
        return "Failed transaction"


a = Account("ACC123", 1234, 5000)
print(a.check_balance(1234))
print(a.withdraw(1000, 1234))
print(a.check_balance(1234))
```

6.  Consider online payment systems such as UPI, Credit Card, Debit Card, Net 
Banking, Wallets. Users do not need to know encryption logic, API calls, or security 
layers.Explain how abstraction simplifies user interaction while preserving internal 
complexity.Write Python code using ABC and abstractmethod for different payment 
methods. (Code + Theory) 

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class UPI(Payment):
    def pay(self, amount):
        return f"Paid {amount} via UPI."

class Card(Payment):
    def pay(self, amount):
        return f"Paid {amount} via Credit Card."

class Wallet(Payment):
    def pay(self, amount):
        return f"Paid {amount} via Wallet."

methods = [UPI(), Card(), Wallet()]
for m in methods:
    print(m.pay(300))
```

B. A credit card transaction dataset shows abnormally high withdrawals for some 
customers. Using Python (Code + Theory): 
1. Detect outliers using both Z-score and IQR 
2. Compare results — do both methods detect the same outliers? Why or why not? 
3. Correct the outliers using any one method: 
a. capping 
b. median replacement 
c. domain-based threshold 
4. Write a note on why we should not blindly delete outliers 
5. Compare and justify: When to use Z-score vs IQR 
a. When to remove vs replace outliers 
b. When visualization is more useful than numeric tables 
c. Give real-world examples for each case.

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.DataFrame({
    "amount": np.concatenate([np.random.normal(50, 10, 190),
                              [500, 700, 900, 1200, 1500]])
})

# Z-score
z = np.abs(stats.zscore(df["amount"]))
z_out = df[z > 3]

# IQR
Q1, Q3 = df["amount"].quantile([0.25, 0.75])
IQR = Q3 - Q1
iqr_out = df[(df["amount"] < Q1 - 1.5*IQR) | (df["amount"] > Q3 + 1.5*IQR)]

print("Z-score outliers:", len(z_out))
print("IQR outliers:", len(iqr_out))

# Fix by capping
lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
df["fixed"] = df["amount"].clip(lower, upper)

print("\nWhy not delete outliers?")
print("Outliers may indicate fraud, system errors, or genuine big transactions.")
print("Deleting them may hide important patterns.")

print("\nWhen to use:")
print("Z-score → normal distributions")
print("IQR → skewed data like finance or spending")
```
